```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Exploratory dataset

```python
salary_data = pd.read_csv("extended_salary_data.csv")

plt.figure(dpi = 150)

x = np.array(salary_data['YearsExperience'])
y = np.array(salary_data['Salary'])
plt.scatter(x,y)
plt.title("Salary vs Experience")
plt.xlabel("No of Years of Experience")
plt.ylabel("Salary")
plt.show()

print("mean of salary : ",np.mean(y))
print("mean of YearExperience : ",np.mean(x))
print("median of salary is : ",np.median(y))
print("median of YearExperience : ",np.median(x))
print("variance of salary is : ",(np.var(y)))
print("variance of YearExperience is : ",np.var(x))
```

---

### Linear Regression Model

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
l_reg = linear_model.LinearRegression()
X = x.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
l_reg.fit(X_train,y_train)
y_pred = l_reg.predict(X_test)

print("R2 score : ",l_reg.score(X_test,y_test))
print("Slope = ",l_reg.coef_)
print("Intercept = ",l_reg.intercept_)
print("MAE score : " , metrics.mean_absolute_error(y_test,y_pred))
print("MSE score : " , metrics.mean_squared_error(y_test,y_pred))

plt.figure(dpi = 150)
plt.plot(X_test,y_pred)
plt.scatter(X_test,y_test)
plt.show()
```
